```python
import pandas as pd
from math import *
```

```python
ori_data = pd.read_table("延页2井标准.txt", sep="\s+", skiprows=6)
ori_data
```

```python
handled_data = ori_data[["#DEPTH", "AC", "DEN"]].round(3)
handled_data = handled_data[
    (handled_data["AC"]>10) & (handled_data["DEN"]>2) & (handled_data["#DEPTH"]>2000)
].reset_index(drop=True)
handled_data
```

```python
handled_data["Vp0"] = 1000000/handled_data["AC"]
```

### 算例

```python
Vp0 = 1000000/333.4776903
Vp90 = (0.7109*(Vp0/1000)+1.6023)*1000
Vp45 = (0.9362*(Vp0/1000)+0.5516)*1000
ts = 333.4776903/(1 - 1.15 * (1/2.348+(1/2.348)**3) / e**(1/2.348))**1.5
Vs0 = 1947.8*log(Vp0)-13746
Vs90 = (0.6479*(Vs0/1000)+1.0489)*1000
C33=(2.348 * Vp0**2)/1000000
C44=(2.348 * Vs0**2)/1000000
C11=(2.348 * Vp90**2)/1000000
C66=(2.348 * Vs90**2)/1000000
C12=C11-2*C66
C13= - C44 + ( (4*(2348**2)*(Vp45**4)*(10**-18)) - 2*2348*(Vp45**2)*(10**-9)*(C11+C33+2*C44) + (C11+C44)*(C33+C44) )**0.5
Ev = C33 - 2*C13**2/(C11+C12)
Eh = (C11-C12) * (C11*C33-2*C13**2+C13*C12) / (C11*C33-C13**2)
Vv = C13 / (C11 + C12)
Vh = (C12*C33 - C13**2) / (C11*C33 - C13**2)
Pp = 1.03 * 9.8 * 3450 / 1000 * 1.18
Pv = 2406*9.8*3450/1000000 
PH_iso = Eh*Vv*(Pv-1*Pp)/(Ev*(1-Vh)) + 1*Pp + Eh*0.96/(1-Vh*Vh) + Eh*Vh*0.35/(1-Vh*Vh)
Ph_iso = Eh*Vv*(Pv-1*Pp)/(Ev*(1-Vh)) + 1*Pp + Eh*0.35/(1-Vh*Vh) + Eh*Vh*0.96/(1-Vh*Vh)
PH_aniso = C13*(Pv-1*Pp)/C33 + 1*Pp + (C12-C13**2/C33)*0.35 + (C11 - C13**2/C33)*0.96
Ph_aniso = C13*(Pv-1*Pp)/C33 + 1*Pp + (C11-C13**2/C33)*0.35 + (C12 - C13**2/C33)*0.96
C11,C12,C13,C33,C44,C66,PH_iso,Ph_iso,PH_aniso,Ph_aniso
```
